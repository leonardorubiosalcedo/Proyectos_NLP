```python
def consultar_comentarios(id_campo):
    instagram = Instagram()
    post1 = instagram.get_media_comments_by_id(id_campo, 1000)
    cadena_comentarios=""
    for a in post1['comments']:
        cadena_comentarios=cadena_comentarios+" "+a.text
    return cadena_comentarios
```

```python
def to_datetime_campo(datestring):  
    
    #date = datetime.strptime(datestring, '%m.%d.%Y')
    #return date.strftime("%Y-%m-%d")
    
    return datetime.strptime(datestring, '%Y-%m-%d')
```

```python
import datetime
def convert_fecha_campo(datos):
    if datos=="Nan":
        return datos
    arreglo = datos.split(",")
    año=arreglo[1].strip()
    arreglo2=arreglo[0].split()
    mes=arreglo2[0]
    dia=arreglo2[1]

    mese={'January':'01', 'February':'02', 'March':'03', 'April':'04', 
              'May' :'05' , 'June':'06' , 'July':'07', 'August':'08' , 'September':'09' , 'October':'10' , 'November':'11' , 'December':'12'}
    mes_campo=mese[mes]
    fecha= datetime.datetime( int(año), int(mes_campo), int(dia))
    return fecha
```

```python
(df["fecha_format"][0] - df["fecha_Respusta_format"][0]).days
```

```python
df["fecha_format"]=df["fecha"].apply(lambda texto: convert_fecha_campo(texto))
df["fecha_Respusta_format"]=df["Fecha_Respusta"].apply(lambda texto: convert_fecha_campo(texto))
```

```python
def diferencia_dias(numero):
    if df["Respuesta"][numero]=="Nan":
        return 0
    else:
        cantidad_dias= (df["fecha_Respusta_format"][numero] - df["fecha_format"][numero]).days
        return cantidad_dias    
```

```python
df["id_registro"]=df.index.copy()
df["dias_para_gestion"]=df["id_registro"].apply(lambda numero: diferencia_dias(numero))
```
