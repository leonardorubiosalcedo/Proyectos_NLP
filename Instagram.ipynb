```python
posts_data = [campo.to_dict() for campo in avvillas_posts]
posts_df = pd.DataFrame(posts_data)
posts_df.head()
```

```python
def nombre_post(id_post):
    instagram = Instagram()
    instagram2 = instagram.Post(id_post)
    instagram2.scrape()
    instagram2.caption
    return instagram2.caption
    #return id_post
```

```python
def consultar_comentarios(id_campo):
    instagram = Instagram()
    post1 = instagram.get_media_comments_by_id(id_campo, 1000)
    cadena_comentarios=""
    for a in post1['comments']:
        cadena_comentarios=cadena_comentarios+" "+a.text
    return cadena_comentarios
```
