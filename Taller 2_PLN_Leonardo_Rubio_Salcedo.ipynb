- Obtener todos los **contenidos** y **etiquetas de título** asociados a los **links** del primer párrafo

```python
# Encabezado, no título
import urllib.request
import bs4 as bs
import numpy as np
url = "https://es.wikipedia.org/wiki/"
biografias = ['Karl_Marx','Mahatma_Gandhi','Leonardo_da_Vinci', "Emmanuel_Macron",
              'Augustin_Robespierre',"Angela_Merkel", "Barack_Obama","Enrique_VIII_de_Inglaterra",
              'Maximilien_Robespierre', 'Servio_Tulio']
```

```python
def busq_html(url):
    request = urllib.request.urlopen(url)
    fuente = request.read()
    request.close()
    arreglo_final_dato=encontrar_dato(fuente)
    return arreglo_final_dato
```

```python
def encontrar_dato(fuente):
    arreglo_link={}
    contador_link=-1
    pagina = bs.BeautifulSoup(fuente, "html.parser")
    primer_parrafo=pagina.find("p") 
    for link in primer_parrafo.find_all("a"):
        contador_link += 1
        ruta_link=link.get("href")
        title=link.get("title")
        contenido=link.contents[0]
        arreglo_link[contador_link]={'ruta' : ruta_link 
                                    , 'title' : title 
                                    , 'contenido' : contenido
                                    }
    titulo=pagina.find("title")
    arreglo_pagina_final={"pagina" : titulo , "links" : arreglo_link}
    return arreglo_pagina_final
```

```python
contador_paginas=-1
arreglo_final_link={}
for paginas in biografias:
    contador_paginas+=1
    url= "https://es.wikipedia.org/wiki/"+paginas
    arreglo_final_link[contador_paginas]=busq_html(url)
```
